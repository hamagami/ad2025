```python
# AnoGAN (GAN を用いた異常検知) — MNIST デモ
# 使い方:
# - 正常クラス (NORMAL_DIGIT) のみで DCGAN を学習
# - 推論時に各入力 x に最も近い潜在ベクトル z* を最適化 (入力固定・z を更新)
# - 異常スコア = (1-ALPHA_SCORE)*残差(L1) + ALPHA_SCORE*特徴距離（D の中間特徴の L2）
# 可視化は英語キャプション、コメントは日本語

import os, random, numpy as np, matplotlib.pyplot as plt, time
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, utils as vutils

# sklearn（未インストールならインストール）
try:
    from sklearn.metrics import roc_auc_score, average_precision_score, roc_curve, auc
except Exception:
    import subprocess; subprocess.call(["pip","-q","install","scikit-learn"])
    from sklearn.metrics import roc_auc_score, average_precision_score, roc_curve, auc

# ================== 設定 ==================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
SEED = 42
BATCH = 128
EPOCHS_GAN = 15          # DCGAN 学習エポック（デモ用）
LR_D = 2e-4
LR_G = 2e-4
BETAS = (0.5, 0.999)

Z_DIM = 100              # 潜在次元（ノイズ z）
NORMAL_DIGIT = 1         # どの数字を「正常」とみなすか (0-9)
VAL_SPLIT = 0.1          # 正常学習データから検証用（しきい値用）に回す割合

# AnoGAN 推論（z 最適化）のハイパラ
Z_OPT_STEPS = 200        # z 最適化の反復回数（多いほど精密・遅い）
Z_LR = 1e-1              # z の最適化学習率（Adam）
LAMBDA_FM = 0.9          # 特徴マッチング重み（損失の重み・残差とのブレンド）
ALPHA_SCORE = 0.9        # 最終スコアでの特徴距離の重み（0〜1）
THRESH_PCT = 99.0        # しきい値（検証正常スコアの分位点）
N_EVAL_PER_CLASS = 300   # テストの各クラスから評価するサンプル数（デモ速度用）

random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

# ================== データ ==================
# GAN は Tanh 出力（[-1,1]）なので [-1,1] 正規化を使用
tfm = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,),(0.5,))])
root = "/content/data"
tr = datasets.MNIST(root, train=True,  download=True, transform=tfm)
te = datasets.MNIST(root, train=False, download=True, transform=tfm)

# 正常/異常インデックス
tr_norm_idx = [i for i,(_,y) in enumerate(tr) if y == NORMAL_DIGIT]
te_norm_idx = [i for i,(_,y) in enumerate(te) if y == NORMAL_DIGIT]
te_anom_idx = [i for i,(_,y) in enumerate(te) if y != NORMAL_DIGIT]

# 正常学習データを bank / val に分割（val はしきい値用。z 最適化が重いので少量が無難）
rng = np.random.default_rng(SEED); rng.shuffle(tr_norm_idx)
n_val = max(1, int(len(tr_norm_idx)*VAL_SPLIT))
val_idx, bank_idx = tr_norm_idx[:n_val], tr_norm_idx[n_val:]

ds_bank = Subset(tr, bank_idx)     # DCGAN 学習（正常のみ）
ds_val  = Subset(tr, val_idx)      # しきい値用（正常のみ）
# テストは速度のためサブサンプリング
te_norm_sub = te_norm_idx[:N_EVAL_PER_CLASS]
te_anom_sub = te_anom_idx[:N_EVAL_PER_CLASS]
ds_tn = Subset(te, te_norm_sub)
ds_ta = Subset(te, te_anom_sub)

def mk_loader(ds, shuffle=False, bs=BATCH):
    return DataLoader(ds, batch_size=bs, shuffle=shuffle, num_workers=2,
                      pin_memory=(DEVICE=="cuda"))
ld_bank = mk_loader(ds_bank, shuffle=True)
ld_val  = mk_loader(ds_val,  shuffle=False, bs=32)
ld_tn   = mk_loader(ds_tn,   shuffle=False, bs=1)   # z 最適化はサンプル単位で実施
ld_ta   = mk_loader(ds_ta,   shuffle=False, bs=1)

print(f"[DATA] normal={NORMAL_DIGIT} | bank={len(ds_bank)} val={len(ds_val)} "
      f"test_norm={len(ds_tn)} test_anom={len(ds_ta)}")

# ================== DCGAN モデル ==================
# 重み初期化（DCGAN 流儀）
def weights_init(m):
    cname = m.__class__.__name__
    if cname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif cname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)

# ジェネレータ: z -> 1x28x28
class Generator(nn.Module):
    def __init__(self, zdim=Z_DIM):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(zdim, 128*7*7),
            nn.BatchNorm1d(128*7*7),
            nn.ReLU(True),
        )
        self.up = nn.Sequential(
            nn.ConvTranspose2d(128, 64, 4, 2, 1),  # 7->14
            nn.BatchNorm2d(64), nn.ReLU(True),
            nn.ConvTranspose2d(64, 32, 4, 2, 1),   # 14->28
            nn.BatchNorm2d(32), nn.ReLU(True),
            nn.Conv2d(32, 1, 3, 1, 1),
            nn.Tanh()                               # [-1,1]
        )
    def forward(self, z):
        h = self.fc(z.view(z.size(0), -1))
        h = h.view(-1,128,7,7)
        return self.up(h)

# 識別器: 1x28x28 -> (本物確率, 中間特徴)
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 4, 2, 1), nn.LeakyReLU(0.2, True),     # 28->14
            nn.Conv2d(32,64, 4, 2, 1), nn.BatchNorm2d(64), nn.LeakyReLU(0.2, True), # 14->7
            nn.Conv2d(64,128,3,1,1), nn.BatchNorm2d(128), nn.LeakyReLU(0.2, True)   # 7->7
        )
        self.head = nn.Conv2d(128, 1, 7, 1, 0)  # 7x7 -> 1x1
    def forward(self, x, return_feat=False):
        feat = self.net(x)                      # [B,128,7,7]
        out = self.head(feat).view(-1,1)        # [B,1]
        if return_feat: return out, feat        # 中間特徴を返す
        return out

G = Generator(Z_DIM).to(DEVICE); G.apply(weights_init)
D = Discriminator().to(DEVICE); D.apply(weights_init)
optG = torch.optim.Adam(G.parameters(), lr=LR_G, betas=BETAS)
optD = torch.optim.Adam(D.parameters(), lr=LR_D, betas=BETAS)
bce = nn.BCEWithLogitsLoss()

# ================== DCGAN 学習（正常のみ） ==================
G.train(); D.train()
for ep in range(1, EPOCHS_GAN+1):
    totD = totG = nimg = 0
    for x,_ in ld_bank:
        x = x.to(DEVICE, non_blocking=True)
        bs = x.size(0); nimg += bs
        # --- Discriminator ---
        z = torch.randn(bs, Z_DIM, device=DEVICE)
        fake = G(z).detach()
        logits_real = D(x)
        logits_fake = D(fake)
        lossD = bce(logits_real, torch.ones_like(logits_real)) + \
                bce(logits_fake, torch.zeros_like(logits_fake))
        optD.zero_grad(); lossD.backward(); optD.step()
        # --- Generator ---
        z = torch.randn(bs, Z_DIM, device=DEVICE)
        fake = G(z)
        logits_fake = D(fake)
        lossG = bce(logits_fake, torch.ones_like(logits_fake))
        optG.zero_grad(); lossG.backward(); optG.step()
        totD += float(lossD)*bs; totG += float(lossG)*bs
    print(f"Epoch {ep:02d}/{EPOCHS_GAN}  D={totD/max(1,nimg):.4f}  G={totG/max(1,nimg):.4f}")

G.eval(); D.eval()

# 参考: 生成画像のグリッドを可視化
z_show = torch.randn(64, Z_DIM, device=DEVICE)
with torch.no_grad():
    fake_show = (G(z_show).cpu()*0.5+0.5)  # 0-1
grid = vutils.make_grid(fake_show, nrow=8, padding=2)
plt.figure(figsize=(6,6)); plt.imshow(grid.permute(1,2,0).numpy(), cmap="gray"); plt.axis("off")
plt.title("DCGAN samples after training"); plt.show()

# ================== AnoGAN: z* 最適化 ==================
# ・x を固定し、z をパラメータとして最適化
# ・損失 = (1-λ)*残差(L1) + λ*特徴距離(L2)  （D の中間特徴を使用）
def anogan_optimize_z(x, steps=Z_OPT_STEPS, z_lr=Z_LR, lambda_fm=LAMBDA_FM, verbose=False):
    # x: [1,1,28,28]
    z = torch.randn(1, Z_DIM, device=DEVICE, requires_grad=True)
    opt = torch.optim.Adam([z], lr=z_lr)
    hist = []
    with torch.enable_grad():
        for t in range(steps):
            x_hat = G(z)
            # 残差: L1
            res = (x_hat - x).abs().mean()
            # 特徴距離: D の中間特徴の L2
            _, fx = D(x, return_feat=True)
            _, fz = D(x_hat, return_feat=True)
            fm = F.mse_loss(fz, fx)
            loss = (1.0 - lambda_fm) * res + lambda_fm * fm
            opt.zero_grad(); loss.backward(); opt.step()
            hist.append([float(res.detach()), float(fm.detach())])
            # z を若干制約（過度な発散を抑える）
            with torch.no_grad():
                z.clamp_(-3.0, 3.0)
    # 最終スコア（残差と特徴距離の線形結合）
    x_hat = G(z).detach()
    _, fx = D(x, return_feat=True)
    _, fz = D(x_hat, return_feat=True)
    res = (x_hat - x).abs().mean().item()
    fm  = F.mse_loss(fz, fx).item()
    score = (1.0 - ALPHA_SCORE) * res + ALPHA_SCORE * fm
    return z.detach(), x_hat, res, fm, score, np.array(hist)

# ================== スコア算出（検証・テスト） ==================
def scores_for_loader(loader, max_items=None):
    scores, res_list, fm_list = [], [], []
    x_list, xh_list = [], []
    t0 = time.time()
    for i, (x,_) in enumerate(loader):
        if (max_items is not None) and (i >= max_items): break
        x = x.to(DEVICE)
        _, x_hat, res, fm, sc, _ = anogan_optimize_z(x, steps=Z_OPT_STEPS)
        scores.append(sc); res_list.append(res); fm_list.append(fm)
        x_list.append(x.cpu()); xh_list.append(x_hat.cpu())
        if (i+1) % 50 == 0:
            dt = time.time()-t0
            print(f"  processed {i+1} samples in {dt:.1f}s")
    X = torch.cat(x_list, 0) if x_list else None
    XH= torch.cat(xh_list,0) if xh_list else None
    return np.array(scores), np.array(res_list), np.array(fm_list), X, XH

print("Scoring validation normals (for threshold)...")
val_scores, val_res, val_fm, _, _ = scores_for_loader(ld_val)

print("Scoring test normals...")
tn_scores, tn_res, tn_fm, Xn, Xn_hat = scores_for_loader(ld_tn)
print("Scoring test anomalies...")
ta_scores, ta_res, ta_fm, Xa, Xa_hat = scores_for_loader(ld_ta)

thr = np.percentile(val_scores, THRESH_PCT)
print(f"Threshold (p{THRESH_PCT:.1f} on validation normals): {thr:.6f}")

# ================== 評価（AUC/AP, ROC） ==================
y_tn = np.zeros_like(tn_scores, dtype=int)
y_ta = np.ones_like(ta_scores, dtype=int)
labels = np.concatenate([y_tn, y_ta])
scores = np.concatenate([tn_scores, ta_scores])
auc_v = roc_auc_score(labels, scores)
ap_v  = average_precision_score(labels, scores)
pred = (scores >= thr).astype(int)
acc = (pred == labels).mean()
print(f"AUC={auc_v:.4f}, AP={ap_v:.4f}, Acc@thr={acc:.4f}")
print(f"Mean components -> Residual (N/A) normal={tn_res.mean():.4f} anomaly={ta_res.mean():.4f} | "
      f"Feature normal={tn_fm.mean():.4f} anomaly={ta_fm.mean():.4f}")

# ================== 可視化 ==================
# 1) スコア・ヒストグラム
plt.figure(figsize=(6,4))
plt.hist(tn_scores, bins=40, alpha=0.6, label="normal")
plt.hist(ta_scores, bins=40, alpha=0.6, label="anomaly")
plt.axvline(thr, linestyle="--", color="red", label=f"threshold={thr:.4f}")
plt.title("AnoGAN anomaly score histogram")
plt.xlabel("score (higher = more anomalous)"); plt.ylabel("count"); plt.legend(); plt.show()

# 2) ROC 曲線
fpr, tpr, _ = roc_curve(labels, scores)
plt.figure(figsize=(5,5))
plt.plot(fpr, tpr, label=f"ROC (AUC={auc(fpr,tpr):.3f})")
plt.plot([0,1],[0,1], linestyle="--")
plt.title("ROC curve"); plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.legend(); plt.grid(True, linestyle=":"); plt.show()

# 3) 再構成ギャラリー（上位異常 / 最も正常）
def show_gallery(X, Xhat, title, k=8, pick="top"):  # pick: "top"=スコア高, "best"=スコア低
    n = min(k, X.shape[0])
    # スコアに応じて並び替え
    if pick == "top":
        order = np.argsort(-scores[-len(X):])[:n]  # 末尾側に該当 X のスコアがあるように合わせて利用（デモ用）
    else:
        order = np.argsort(scores[-len(X):])[:n]
    Xs, Xhs = X[order], Xhat[order]
    resid = (Xs - Xhs).abs()
    plt.figure(figsize=(4*n, 4))
    for i in range(n):
        plt.subplot(1, 3*n, 3*i+1); plt.imshow((Xs[i,0]*0.5+0.5), cmap="gray"); plt.axis("off"); plt.title("Original")
        plt.subplot(1, 3*n, 3*i+2); plt.imshow((Xhs[i,0]*0.5+0.5), cmap="gray"); plt.axis("off"); plt.title("Reconstruction")
        plt.subplot(1, 3*n, 3*i+3); plt.imshow(resid[i,0], cmap="gray"); plt.axis("off"); plt.title("Residual |x−x̂|")
    plt.suptitle(title); plt.show()

# 異常側の上位を表示
if Xa is not None:
    k = min(8, Xa.shape[0])
    show_gallery(Xa, Xa_hat, "Top-scored anomalies", k=k, pick="top")
# 正常側の最も正常を表示
if Xn is not None:
    k = min(8, Xn.shape[0])
    show_gallery(Xn, Xn_hat, "Most normal samples", k=k, pick="best")

print("""
Notes:
- This is a minimal AnoGAN demo: DCGAN trained on normal class; per-sample z is optimized at test time.
- Score = (1-α)*Residual(L1) + α*Feature distance (Discriminator mid-layer MSE).
- Increase Z_OPT_STEPS for better z-matching (but slower). Tune LAMBDA_FM and ALPHA_SCORE for balance.
- For larger images / speed, consider f-AnoGAN (encoder for one-shot z) instead of iterative optimization.
""")
```
